```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings ('ignore')
```

```python
f=requests.get('https://www.practo.com/delhi/doctors-for-skin-disease-treatment').text
```

```python
f
```

```python
soup=BeautifulSoup(f,'html.parser')
```

```python
soup
```

```python
soup.prettify()
```

```python
s=soup.select('div.container')
```

```python
[i.text for i in soup.find_all('h2')]
```

```python
fees=[i.text for i in soup.select(' div.uv2-spacer--xs-top ') ]
```

```python

treatment=pd.DataFrame()

temp_data=pd.DataFrame(columns=['Dr.name/clinic','cunsultant_fee'])
clinic=[i.text for i in soup.find_all('h2')]
clinic=[clinic[i]for i in range(len(clinic)-9)]
fees=[i.text for i in soup.select(' div.uv2-spacer--xs-top ') ]
fee=[]
for i in range(len(fees)):
    if i>=2 and i %2==0:
        fee.append(fees[i].split()[0])

for i in range(len(clinic)):
    db={}
    temp_data=pd.DataFrame(columns=['Dr.name/clinic','cunsultant_fee'])
    db['Dr.name/clinic']=clinic[i]
    db['cunsultant_fee']=fee[i]

    df=temp_data.append(db,ignore_index=True)
    treatment=treatment.append(df,ignore_index=True)
    
```
